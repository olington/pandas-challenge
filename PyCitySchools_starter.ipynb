```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df
```

```python
#Calculate the total number of schools
school_total_number = school_data_complete_df["school_name"].nunique()
school_total_number
```

```python
# Calculate the total number of students
student_number = school_data_complete_df["Student ID"].count()
student_number
```

```python
#Calculate the total budget
budget_total = school_data_df["budget"].sum()
budget_total
```

```python
#Calculate the average maths score
Avg_score_maths = school_data_complete_df["maths_score"].mean()
Avg_score_maths 
```

```python
#Average reading score
Avg_score_reading = school_data_complete_df["reading_score"].mean()
Avg_score_reading
```

```python
#Percentage of students with a passing maths score above 50 (pass mark is 50)
#first calculate the number with score above 50
maths_pass_count = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)].count()["student_name"]
maths_pass_percent = maths_pass_count / float(student_number ) * 100
```

```python
#Percentage of students with a passing reading score above 50
#first calculate the number with reading score above 50
reading_pass_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)].count()["student_name"]
reading_pass_percent = maths_pass_count / float(student_number ) * 100
```

```python
#Percentage of students who passed maths and reading overall
#first calculate the number with overall score above 50
maths_reading__pass_count = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50) 
                                                  & (school_data_complete_df["reading_score"] >= 50)].count()["student_name"]
overall_pass_percentage = maths_reading__pass_count / float(student_number ) * 100
```

```python
#Clean, reorganize and create dataframe
area_summary = pd.DataFrame({"Total number of Schools": [school_total_number], 
                                 "Total number of Students": [student_number], 
                                 "Total Budget": [budget_total],
                                 "Average Maths Score": [Avg_score_maths ], 
                                 "Average Reading Score": [Avg_score_reading],
                                 "Percentage Passing Maths": [maths_pass_percent],
                                 "Percentage Passing Reading": [reading_pass_percent],
                                 "Percentage Overall Passing": [overall_pass_percentage]})

Area_summary = area_summary[["Total number of Schools",
                             "Total number of Students",
                             "Total Budget",
                             "Average Maths Score", 
                             "Average Reading Score",
                             "Percentage Passing Maths",
                             "Percentage Passing Reading",
                             "Percentage Overall Passing"]]

# Formatt as needed in display
Area_summary["Total number of Students"] = Area_summary["Total number of Students"].map("{:,}".format)
Area_summary["Total Budget"] = Area_summary["Total Budget"].map("${:,.2f}".format)

# Display DataFrame
Area_summary
```

```python
#get school Names and type
school_type = school_data_df.set_index(["school_name"])["type"]
school_type
```

## School Summary

```python
#Get the number of students per school
student_count_school = school_data_complete_df["school_name"].value_counts()
```

```python
#get the total School Budget
each_school_budget = school_data_complete_df.groupby(["school_name"])["budget"].mean()
```

```python
#get the budget per Student
budget_per_student = each_school_budget / student_count_school
```

```python
#get the Average Maths Score
Average_maths_score = school_data_complete_df.groupby(["school_name"])["maths_score"].mean()
```

```python
#get the Average reading Score
Average_reading_score = school_data_complete_df.groupby(["school_name"])["reading_score"].mean()
```

```python
#To get the percentage number of students who passed maths and reading, first create separate dataframes of school subject scores
school_maths_score = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)]

#repeat for reading
school_reading_score = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)]

# Get in a dataframe the number of students who passed both reading and maths too.
school_maths_and_reading = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)
                            & (school_data_complete_df["maths_score"] >= 50)]
```

```python
#Now, get the percentage of maths pass rate in schools
per_school_passing_maths = school_maths_score.groupby(["school_name"]).size() / student_count_school * 100
```

```python
#get the percentage of reading pass rate in schools
per_school_passing_reading = school_reading_score.groupby(["school_name"]).size() / student_count_school * 100
```

```python
#get the overall percentage of maths and reading pass rate in schools
overall_passing_rate = school_maths_and_reading.groupby(["school_name"]).size() / student_count_school * 100
```

```python
# Organise and convert to dataframe
School_summary = pd.DataFrame({"School Type": school_type,
                             "Total Students": student_count_school,
                             "Total School Budget": each_school_budget,
                             "Per Student Budget": budget_per_student,
                             "Average Maths Score": Average_maths_score,
                             "Average Reading Score": Average_reading_score,
                             "Percent Passing Maths": per_school_passing_maths,
                             "Percent Passing Reading": per_school_passing_reading,
                             "Percent Overall Passing": overall_passing_rate})

#formatt dataframe
schools_summary = School_summary[["School Type",
                                 "Total Students",
                                 "Total School Budget",
                                 "Per Student Budget",
                                 "Average Maths Score",
                                 "Average Reading Score", 
                                 "Percent Passing Maths",
                                 "Percent Passing Reading", 
                                 "Percent Overall Passing"]]

schools_summary["Total School Budget"] = schools_summary["Total School Budget"].map("${:,.2f}".format)
schools_summary["Per Student Budget"] = schools_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
schools_summary
```

## Top Performing Schools by percentage Overall Passing

```python
#get the top 5 performing schools by sorting
top_schools = schools_summary.sort_values(["Percent Overall Passing"], ascending=False)
top_schools.head(5)
```

## Bottom Performing Schools by percentage Overall Passing

## Maths Scores by Year

```python
# create a series of the scores by year levels using conditionals
year_nine_data = school_data_complete_df[(school_data_complete_df["year"] == 9)]
year_ten_data = school_data_complete_df[(school_data_complete_df["year"] == 10)]
year_eleven_data = school_data_complete_df[(school_data_complete_df["year"] == 11)]
year_twelve_data = school_data_complete_df[(school_data_complete_df["year"] == 12)]

# Groupby each by school name and maths score
year_nine_maths_scores = year_nine_data.groupby(["school_name"])["maths_score"].mean()
year_ten_maths_scores = year_ten_data.groupby(["school_name"])["maths_score"].mean()
year_eleven_maths_scores = year_eleven_data.groupby(["school_name"])["maths_score"].mean()
year_twelve_maths_scores = year_twelve_data.groupby(["school_name"])["maths_score"].mean()

# Get all the years into a single DataFrame
maths_scores_by_year = pd.DataFrame({"Year 9": year_nine_maths_scores,
                                     "Year 10": year_ten_maths_scores,
                                     "Year 11": year_eleven_maths_scores,
                                     "Year 12": year_twelve_maths_scores})

#arrange
Maths_scores_by_year = maths_scores_by_year[["Year 9", "Year 10", "Year 11", "Year 12"]]

maths_scores_by_year
```

## Reading Score by Year

```python
# create a series of the scores by year levels using conditionals
year_nine_data = school_data_complete_df[(school_data_complete_df["year"] == 9)]
year_ten_data = school_data_complete_df[(school_data_complete_df["year"] == 10)]
year_eleven_data = school_data_complete_df[(school_data_complete_df["year"] == 11)]
year_twelve_data = school_data_complete_df[(school_data_complete_df["year"] == 12)]

# Groupby each by school name and reading score
year_nine_reading_scores = year_nine_data.groupby(["school_name"])["reading_score"].mean()
year_ten_reading_scores = year_ten_data.groupby(["school_name"])["reading_score"].mean()
year_eleven_reading_scores = year_eleven_data.groupby(["school_name"])["reading_score"].mean()
year_twelve_reading_scores = year_twelve_data.groupby(["school_name"])["reading_score"].mean()

# Get all the years into a single DataFrame
reading_scores_by_year = pd.DataFrame({"Year 9": year_nine_reading_scores,
                                     "Year 10": year_ten_reading_scores,
                                     "Year 11": year_eleven_reading_scores,
                                     "Year 12": year_twelve_reading_scores})

# Arrange
Reading_scores_by_year = reading_scores_by_year[["Year 9", "Year 10", "Year 11", "Year 12"]]

Reading_scores_by_year
```

## Scores by School Spending

```python
#creating bins 
bin_limits = [0, 585, 630, 645, 680]
bin_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# copy school summary dataframe to get to "Per Student Budget" 
schools_summary_copy_df = schools_summary
```

```python
#create a new column to categorise spending while sorting "per student budget" spending into bins created
schools_summary_copy_df["Spending Ranges (Per Student)"]  = pd.cut(budget_per_student, bin_limits, labels=bin_labels, right=False)
```

```python
#  Calculate averages for the desired columns. 
spending_by_avg_maths = schools_summary_copy_df.groupby(["Spending Ranges (Per Student)"])["Average Maths Score"].mean()
spending_by_avg_reading = schools_summary_copy_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_maths = schools_summary_copy_df.groupby(["Spending Ranges (Per Student)"])["Percent Passing Maths"].mean()
spending_passing_reading = schools_summary_copy_df.groupby(["Spending Ranges (Per Student)"])["Percent Passing Reading"].mean()
overall_passing_spending = schools_summary_copy_df.groupby(["Spending Ranges (Per Student)"])["Percent Overall Passing"].mean()
```

```python
# create a new dataframe and round to 2 where necessary
spending_summary = pd.DataFrame({"Avg Maths Score" : spending_by_avg_maths.round(2), "Avg Reading Score" : spending_by_avg_reading.round(2), "Percentage Passing Maths" : spending_passing_maths.round(2), "Percentage Passing Reading" : spending_passing_reading.round(2), "Percentage Overall Passing" : overall_passing_spending.round(2)})


spending_summary = spending_summary[["Avg Maths Score", 
                                     "Avg Reading Score", 
                                     "Percentage Passing Maths",
                                     "Percentage Passing Reading",
                                     "Percentage Overall Passing"]]


spending_summary
```

## Scores by School Size

```python
#creating bins 
bin_limits = [0, 1000, 2000, 5000]
bin_labels = ["<$1000", "$1000-2000", "$2000-5000"]
```

```python
#create a new column to categorise size while sorting "Total students" into bins created
schools_summary_copy_df["Ranges (School size)"]  = pd.cut(student_count_school, bin_limits, labels=bin_labels, right=False)
```

```python
#  Calculate averages for the desired columns. 
size_by_avg_maths = schools_summary_copy_df.groupby(["Ranges (School size)"])["Average Maths Score"].mean()
size_by_avg_reading = schools_summary_copy_df.groupby(["Ranges (School size)"])["Average Reading Score"].mean()
size_passing_maths = schools_summary_copy_df.groupby(["Ranges (School size)"])["Percent Passing Maths"].mean()
size_passing_reading = schools_summary_copy_df.groupby(["Ranges (School size)"])["Percent Passing Reading"].mean()
overall_passing_size = schools_summary_copy_df.groupby(["Ranges (School size)"])["Percent Overall Passing"].mean()
```

```python
# create a new dataframe and round to 2 where necessary
size_summary = pd.DataFrame({"Avg Maths Score" : size_by_avg_maths,
                                 "Avg Reading Score" : size_by_avg_reading,
                                 "Percentage Passing Maths" : size_passing_maths,
                                 "Percentage Passing Reading" : size_passing_reading,
                                 "Percentage Overall Passing" : overall_passing_size})


size_summary = size_summary[["Avg Maths Score", 
                                     "Avg Reading Score", 
                                     "Percentage Passing Maths",
                                     "Percentage Passing Reading",
                                     "Percentage Overall Passing"]]


size_summary
```
